```python
import pandas as pd
Temp2015 = pd.read_excel('Temp_2015.xlsx')
Temp2016 = pd.read_excel('Temp_2016.xlsx')
Temp2017 = pd.read_excel('Temp_2017.xlsx')
Temp2018 = pd.read_excel('Temp_2018.xlsx')

TempAll = [Temp2015,Temp2016,Temp2017,Temp2018]

append_df = pd.concat(TempAll)
append_df.head(2)

# append_df.to_excel('TempAll.xlsx',index=False)
```

```python
append_df['Date']=pd.PeriodIndex(year=append_df['Year'],month=append_df['Month'],day=append_df['Day'],freq='D')
append_df.head(2)
```

```python
Temperature = pd.DataFrame(append_df, columns = ['Date','Year','Month','Day','Hour','Dry Bulb (F)'])
Temperature.head()

Temperature.to_excel('TempAll.xlsx',index=False)
```
