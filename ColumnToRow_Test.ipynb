```python
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/Daisy/Documents/PyTest/test/xx/Power.csv')
data.head(2)
```

```python
data = data.set_index(['TDSPCODE','ACCOUNT NO. ','SERVICEPOINTID','OPERATINGDATE'])
data = data.stack()
```

```python
data.index=data.index.rename('Hour',level=4)

data.name='Power(KW)'

data=data.reset_index()
data.head(2)
```

```python
#data.to_csv('ConvertedPower.csv')
```

```python
data['DATE'] = data['OPERATINGDATE'].map(lambda x:x.split(' ')[0])
data.head(2)
```

```python
df = data.drop(columns=['OPERATINGDATE','TDSPCODE','ACCOUNT NO. ','SERVICEPOINTID'])
df.head(10)
```

```python
# df.dtypes
df['Hour'] = df['Hour'].astype('str').str[1:].astype(int)
df.head(10)
# df.dtypes
```

```python
df['Hour_new']=df['Hour'].map(lambda x:(x-1))
df.head()
```

```python
Power = pd.DataFrame(df, columns= ['DATE','Hour_new','Power(KW)'])
Power.head()
```

```python
Power.columns = ['Date','Hour','Power(KW)']
Power.head()
```

```python
import datetime

Power.dtypes
```

```python
Temperature = pd.read_excel('C:/Users/Daisy/Documents/PyTest/test/xx/Temp/TempAll.xlsx')
Temperature['DateHour'] = Temperature['Date'].astype(str) + ' H' + Temperature['Hour'].astype(str)

Temperature.head(2)
```

```python
Combine = pd.merge (Temperature,
                   Power,
                   on = 'DateHour',
                   how = 'inner')
Combine.head()
```

```python
Combine_new = pd.DataFrame(Combine, columns=['Date_x','Year','Month','Day','Hour_x','Dry Bulb (F)','Power(KW)'])
Combine_new.head()
```

```python
Final = Combine_new.rename(columns={'Date_x':'Date',
                                   'Hour_x':'Hour'})
Final.head()
```
